```python
new_df = df.dropna()
```

```python
data = new_df
```

```python
data.describe()
```

```python
for col in data.columns[1:]:
    data = data.drop(index=data[data[col]==-100].index)
```

```python
data.columns
```

```python
data
```

```python
data.describe()
```

```python
data['time'] = pd.to_datetime(data['time'], format="%Y-%m-%d %H:%M:%S")
```

```python
data['Hour'] = data['time'].dt.hour
hour = data.groupby('Hour').mean().reset_index()
```

```python
hour
```

```python
import matplotlib.pyplot as plt
```

```python
# fig, axes = plt.subplots(4, 1, figsize=(8, 20))
# axes[0].bar(hour['Hour'], hour['Users'], color='r')
# axes[1].bar(hour['Hour'], hour['Sessions'], color='g')
# axes[2].bar(hour['Hour'], hour['NewVisitors'], color='b')
# axes[3].bar(hour['Hour'], hour['PageViews'], color='c')
fig, axes = plt.subplots(len(list(data.columns)[1:]), 1, figsize=(8, 300))
for i, title in enumerate(list(data.columns)[1:]):
    axes[i].set_title(title)
    axes[i].bar(hour['Hour'], hour[title])
plt.show()
```
